```python
# import ve read işlemlerini yaptım
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
normal_season = pd.read_csv("Normalseason.csv", encoding="Latin1", sep=';')
playoffs_season = pd.read_csv("Playoffs.csv", encoding="Latin1", sep=";")
normal_season
```

```python
# bazı oyuncuların 3 takımı olduğu için ilk satırların geçen sezondan kaldığını düşünerek, ilk satırlarını siliyorum
# Oyuncuları Rk numarasına göre gruplayın
rk_groups = normal_season.groupby('Rk')
filtered_rows = rk_groups.apply(lambda x: x.iloc[1:] if len(x) > 1 else x)
cleared_normal_season = filtered_rows.reset_index(drop=True)
cleared_normal_season
```

```python
# her oyncunun son takını belirleyip isim üzerinden join yapalım
players_last_teams = cleared_normal_season.groupby('Player').agg({'Tm': 'last'}).reset_index()
players_last_teams
```

```python
# önce oyuncuları gruplayıp, sonrasında iki sezonda da oynamış oyuncuları seçelim
# normal sezonu grupladım
numeric_columns = cleared_normal_season.select_dtypes(include='number').columns
sum_columns = ["G", "GS"]
mean_columns = cleared_normal_season.columns[8:]

grouped_sum = cleared_normal_season.groupby("Player")[sum_columns].sum()
grouped_mean = cleared_normal_season.groupby("Player")[mean_columns].mean()
edited_normal_season = pd.concat([grouped_sum, grouped_mean], axis=1).reset_index()
edited_normal_season = edited_normal_season.round(2)

# normal sezonu takımlara join ettim
edited_normal_season = pd.merge(edited_normal_season, players_last_teams, on="Player")
edited_normal_season
```

```python
# playoffs sezonunu grupladım
grouped_sum = playoffs_season.groupby("Player")[sum_columns].sum()
grouped_mean = playoffs_season.groupby("Player")[mean_columns].mean()
edited_playoffs_season = pd.concat([grouped_sum, grouped_mean], axis=1).reset_index()
edited_playoffs_season = edited_playoffs_season.round(2)

# playoff sezonunu takımlara join ettim
edited_playoffs_season = pd.merge(edited_playoffs_season, players_last_teams, on="Player")
edited_playoffs_season
```

```python
# iki sezonda da oynamış olan oyuncuları filtreledim. buradan sonra bu liste ile devam edeceğim
edited_normal_season = edited_normal_season[edited_normal_season["Player"].isin(edited_playoffs_season["Player"])]
edited_normal_season
```

```python
# iki sezonda da oynamış olan oyuncuların, iki ayrı sezondaki eff puanlarını hesapladım
edited_normal_season["EFF"] = (edited_normal_season["PTS"] + edited_normal_season["TRB"] + edited_normal_season["AST"] + edited_normal_season["STL"] + edited_normal_season["BLK"]) - (edited_normal_season["FGA"] - edited_normal_season["FG"] + edited_normal_season["FTA"] - edited_normal_season["FT"] + edited_normal_season["TOV"] + edited_normal_season["PF"])
eff_n = edited_normal_season[["Player", "EFF", "Tm"]].sort_values(by="EFF", ascending=False)
eff_n
```

```python
# playoffs versiyonunu da hesapladım
edited_playoffs_season["EFF"] = (edited_playoffs_season["PTS"] + edited_playoffs_season["TRB"] + edited_playoffs_season["AST"] + edited_playoffs_season["STL"] + edited_playoffs_season["BLK"]) - (edited_playoffs_season["FGA"] - edited_playoffs_season["FG"] + edited_playoffs_season["FTA"] - edited_playoffs_season["FT"] + edited_playoffs_season["TOV"] + edited_playoffs_season["PF"])
eff_p = edited_playoffs_season[["Player", "EFF", "Tm"]].sort_values(by="EFF", ascending=False)
eff_p
```
